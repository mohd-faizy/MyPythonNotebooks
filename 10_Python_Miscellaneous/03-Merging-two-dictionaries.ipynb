### Using the `|` Operator (Python 3.9+)

This is the simplest and most modern way to merge dictionaries:

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

dict1.update(dict2)

print(dict1)
```

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

dict2.update(dict1)

print(dict2)
```

### Using Dictionary Unpacking (`**`)

This method works in Python 3.5 and above:

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```

### Using the `copy()` and `update()` Methods

This method first copies one dictionary and then updates it with the other:

```python
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}

merged_dict = dict1.copy()
merged_dict.update(dict2)

print(merged_dict)
```
